# Visualize the predictions

```python
def visualize_model_predictions(model, x_test, y_test, title_string):
    y_pred = model.predict(x_test)
    
    figure = plt.figure(figsize=(20, 8))
    for i, index in enumerate(np.random.choice(x_test.shape[0], size=32, replace=False)):
        ax = figure.add_subplot(4, 8, i+1, xticks=[], yticks=[])
        
        ax.imshow(np.squeeze(x_test[index]), cmap='gray')
        predicted_index = np.argmax(y_pred[index])
        true_index = np.argmax(y_test[index])
        
        ax.set_title('{}({})'.format(dataset_labels[predicted_index], 
                                     dataset_labels[true_index]),
                                     color=('green' if predicted_index==true_index else 'red'))
            
    figure.suptitle("%s results:" %title_string, fontsize=25)
    
    
    
visualize_model_predictions(model_baseline, x_test, y_test, 'Test')
```

# Add Callbacks

Callbacks - To reduce overfitting, we will want the model to stop training if


  The validation loss keeps on decreasing consistently (For eg. 10 times in a row) 


We also want to see how the validation error is changing.

```python
class plot_train_progress(keras.callbacks.Callback):
    
    def __init__(self, plot_every_n_epochs=5, plot_fraction=0.9):
        self.plot_frequency = plot_every_n_epochs
        self.plot_fraction = plot_fraction
    
    def on_train_begin(self, logs={}):
        self.train_losses = []
        self.val_losses = []
    
    def on_epoch_end(self, epoch, logs={}):
        print('epoch =', epoch)
        if epoch % self.plot_frequency == 0:
            nr_values_to_plot = int(self.plot_fraction * len(self.train_losses))
            if nr_values_to_plot > 1:
                t = list(range(nr_values_to_plot))
                plt.plot(t, self.train_losses[-nr_values_to_plot: ], 'b')
                plt.plot(t, self.val_losses[-nr_values_to_plot: ], 'g')
                plt.xlabel('# Batches')
                plt.ylabel('Validation Loss')
                #plt.savefig('train_progress.jpg')
                plt.show()
                
    def on_batch_end(self, batch, logs={}):
        self.train_losses.append(logs.get('loss'))
        self.val_losses.append(logs.get('val_loss'))


early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=20, verbose=1, mode='min')
filepath='/home/saibharadwaj/Downloads/Ast/ComputerVisionProjects/Fashion_MNIST/model4.weights.best.hdf5'
checkpointer = ModelCheckpoint(filepath=filepath, verbose = 1, save_best_only=True)
callbacks_list = [checkpointer, plot_train_progress(5, 0.9), early_stopping]
```

```python
model.fit(x_train, y_train, batch_size=128, epochs=25, validation_data=(x_valid, y_valid), 
          callbacks=callbacks_list)
```

```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img

standardize_features = False
train_data_gen = ImageDataGenerator(featurewise_center=standardize_features, 
                                    featurewise_std_normalization=standardize_features, 
                                    rotation_range=10, 
                                    width_shift_range=0.1, height_shift_range=0.1, 
                                    shear_range=0.05, zoom_range=0.03, horizontal_flip=False
                                   )

# this is the augmentation configuration we will use for testing:
validation_datagen = ImageDataGenerator(featurewise_center=standardize_features, 
                                        featurewise_std_normalization=standardize_features)
test_datagen = ImageDataGenerator(featurewise_center=standardize_features, 
                                  featurewise_std_normalization=standardize_features)

# compute quantities required for featurewise normalization
# (std, mean, and principal components if ZCA whitening is applied)
train_data_gen.fit(x_train)
validation_datagen.fit(x_valid)
test_datagen.fit(x_test)
```

```python
def plot_batch(images, batch_id):
    assert len(images) >= 30
    figure = plt.figure(figsize=(20, 3.5))
    
    for i, index in enumerate(np.random.choice(images.shape[0], size=30, replace=False)):
        ax = figure.add_subplot(3, 10, i + 1, xticks=[], yticks=[])
        ax.imshow(np.squeeze(images[index]), cmap = 'gray')
    print("Augmented Images, batch %d:" %(batch_id+1))
    plt.show()
    
batches_to_plot, i = 3, 0
    
for x_batch, y_batch in train_data_gen.flow(x_train, y_train, batch_size=32):
    plot_batch(x_batch, i)
    i += 1
    if i == batches_to_plot: 
        break
```
